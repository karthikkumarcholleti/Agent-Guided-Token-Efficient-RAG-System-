```python
from transformers import DistilBertTokenizerFast, DistilBertForSequenceClassification
import torch

tokenizer = DistilBertTokenizerFast.from_pretrained("./distilbert-query-classifier")
model = DistilBertForSequenceClassification.from_pretrained("./distilbert-query-classifier")
model.eval() 
```

```python
def classify_query(query):
    inputs = tokenizer(query, return_tensors="pt", truncation=True, padding=True, max_length=128)
    with torch.no_grad():
        outputs = model(**inputs)
        pred = torch.argmax(outputs.logits, dim=1).item()
    mapping = {0: "Simple", 1: "Follow-Up", 2: "Complex"}
    return mapping[pred]

query_type = classify_query("What is the status of the previous ticket?")
print(query_type)
```

```python
print(classify_query("was it reviewed ?"))
```

```python
questions_s = ['What is the deadline for submitting the quarterly report?', 
             'Who is responsible for the new project launch?',
             'How do I reset my corporate email password?',
             'What are the core working hours for the team?',
             'Which document contains the company’s leave policy?']

for question in questions_s:
    print(classify_query(question))
```

```python
question_f= ['Did it get approved?',
             'Can you clarify the last point?',
            'What about the budget for the marketing team?',
            'Has she responded yet?',
            'Should we schedule the meeting for next week or the week after?']

for question in question_f:
    print(classify_query(question))
```
